```python
import cv2
```

```python
img = cv2.imread('image.jpeg')
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img_fix = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fix)
```

```python
img_gray = cv2.imread('image.jpeg',cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray,cmap = 'gray')
```

```python
img_new = cv2.resize(img_fix,(350,150))
plt.imshow(img_new)
```

```python
width_ratio = 1.25
height_ratio = 0.8
```

```python
img_ratio = cv2.resize(img_fix,
                       (0,0),
                       img_fix,
                       width_ratio,
                       height_ratio
)
```

```python
plt.imshow(img_ratio)
```

```python
img_hor = cv2.flip(img_fix,0)
plt.imshow(img_hor)
```

```python
img_ver = cv2.flip(img_fix,1)
plt.imshow(img_ver)
```

```python
img_both = cv2.flip(img_fix,-1)
plt.imshow(img_both)
```

```python
black_img = np.zeros((512,512,3),dtype = np.int16)
```

```python
cv2.circle(img = black_img,
           center = (400,100),
           radius = 50,
           color = (255,0,0),
           thickness = 8
)
plt.imshow(black_img)
```

```python
cv2.circle(img = black_img,
           center = (200,200),
           radius = 50,
           color = (0,255,0),
           thickness = 8
)
plt.imshow(black_img)
```

```python
cv2.circle(img = black_img,
           center = (100,400),
           radius = 50,
           color = (0,0,255),
           thickness = -1
)
plt.imshow(black_img)
```

```python
cv2.rectangle(img = black_img,
           pt1 = (200,200),
           pt2 = (100,400),
           color = (255,255,0),
           thickness = 8
)
plt.imshow(black_img)
```

```python
cv2.rectangle(img = black_img,
           pt1 = (200,200),
           pt2 = (400,100),
           color = (255,255,0),
           thickness = 8
)
plt.imshow(black_img)
```

```python
vertices = np.array([[100,400],[200,200],[400,100]],np.int32)
pts = vertices.reshape(-1,1,2)
cv2.polylines(black_img,
           [pts],
            isClosed = True,  
           color = (255,0,255),
           thickness = 8
)
plt.imshow(black_img)
```

```python
cv2.fillPoly(black_img,
           [pts], 
           color = (255,255,255)
)
plt.imshow(black_img)
```

```python
cv2.line(black_img,
        pt1 = (0,0),
        pt2 = (500,500),
        color = (0,255,0),
        thickness = 8)
plt.imshow(black_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_img,
           text = 'Euclid',
           org = (100,100),
           fontFace = font,
           fontScale = 3,
           color = (255,0,255),
           thickness = 8,
           lineType = cv2.LINE_AA)
plt.imshow(black_img)
```
